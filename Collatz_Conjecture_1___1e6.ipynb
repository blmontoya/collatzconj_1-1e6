```python
from matplotlib import pyplot as plt
```

```python
def collatz(x):
  a = [x]
  while x != 1:
    if x % 2 == 0:
      x = x/2
    else:
      x = 3*x + 1
    a.append(x)
  return(a)

x = []
y = []

for i in range(1,1000000):
  a = collatz(i)
  x.append(i)
  y.append(len(a)-1)
plt.plot(x,y,".")
plt.show

number = 1
most_iter = 1

for j in range(len(y)):
  if y[j-1] > most_iter:
    most_iter = y[j-1]
    number = x[j-1]

print("        Number: " + str(number) +  "     Most iterations: " + str(most_iter) + "\n")
```
